# Naxos Music Library

In this notebook I'll clean the data scraped previously. One big thing I want to do is get each movie where a piece was used to be on it's own line.

```python
year = 2019
month = 10
```

```python
import csv
```

```python
with open("2019-10-untidy.csv", "r") as file:
    reader = csv.reader(file)
    untidy = list(reader)
```

```python
tidy = [["composer", "piece", "piece_url", "piece_item_code", "movie"]]
```

```python
untidy[0]
```

```python
for row in untidy[1:]:
    composer, piece, piece_url, piece_item_code = row[0:4]
    movies = row[-1].split("?;")
    for movie in movies:
        tidy.append([composer, piece, piece_url, piece_item_code, movie])
        
    
```

```python
tidy
```

```python
out_data_file = "-".join([str(year), str(month), "tidy"])+".csv"
with open(out_data_file, "w") as file:
    csv.writer(file, delimiter=',').writerows(tidy)
```
